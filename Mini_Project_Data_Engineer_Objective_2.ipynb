```python
import pandas as pd
import matplotlib.pyplot as plt

headers = ['Product Name', 'Unit Price', 'Total Quantity Sold', 'Percentage of Total Quantity Sold', 'Average Sales', 'Total Sales', 'Percentage of Total Sales']
df = pd.read_csv("/content/Case Study.csv", names=headers)
```

```python
# sort the dataframe by Total Quantity Sold in descending order
df = df.sort_values("Total Sales", ascending=False)

df
```

```python
# Plot the data to visualize the correlation
plt.scatter(df['Total Quantity Sold'], df['Total Sales'])
plt.xlabel('Total Quantity Sold')
plt.ylabel('Total Sales')
plt.title('Correlation between Total Quantity Sold and Total Sales')
plt.show()

# Calculate the Pearson correlation coefficient
correlation = df['Total Quantity Sold'].corr(df['Total Sales'])
print("The Pearson correlation coefficient between Total Quantity Sold and Total Sales is:", correlation)
```

```python
# Plot the data to visualize the correlation
plt.scatter(df['Total Quantity Sold'], df['Unit Price'])
plt.xlabel('Total Quantity Sold')
plt.ylabel('Unit Price')
plt.title('Correlation between Total Quantity Sold and Unit Price')
plt.show()

# Calculate the Pearson correlation coefficient
correlation = df['Total Quantity Sold'].corr(df['Unit Price'])
print("The Pearson correlation coefficient between Total Quantity Sold and Unit Price is:", correlation)
```

```python
# Plot the data to visualize the correlation
plt.scatter(df['Unit Price'], df['Average Sales'])
plt.xlabel('Unit Price')
plt.ylabel('Average Sales')
plt.title('Correlation between Unit Price and Average Sales')
plt.show()

# Calculate the Pearson correlation coefficient
correlation = df['Unit Price'].corr(df['Average Sales'])
print("The Pearson correlation coefficient between Unit Price and Average Sales is:", correlation)
```

```python
# Create a histogram to visualize the distribution of Total Quantity Sold for each product
df.hist(column='Total Quantity Sold', bins=30, grid=False, color='#86bf91', zorder=2, rwidth=0.9)

# Add labels and title to the plot
plt.xlabel("Total Quantity Sold")
plt.ylabel("Frequency")
plt.title("Distribution of Total Quantity Sold for each Product")

# Show the plot
plt.show()
```

```python
# Group the data by product name and calculate the sum of Total Quantity Sold for each product
grouped_df = df.groupby("Product Name").sum()

# Create a bar plot to visualize the sum of Total Quantity Sold for each product
grouped_df.plot(kind='bar', y='Total Quantity Sold', color='#86bf91', figsize=(15,5), legend=False)

# Add labels and title to the plot
plt.xlabel("Product Name")
plt.ylabel("Total Quantity Sold")
plt.title("Sum of Total Quantity Sold for each Product")

# Show the plot
plt.show()
```

```python
# sort the dataframe by Total Sales in descending order
df = df.sort_values("Total Sales", ascending=False)

# plot the bar plot
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(df["Product Name"], df["Total Sales"])
ax.set_xlabel("Product Name")
ax.set_ylabel("Total Sales")

# rotate the x-axis labels to make them visible
plt.xticks(rotation=90)

# adjust the size of the plot to make the product names more visible
plt.tight_layout()

# show the plot
plt.show()
```

```python
# sort the dataframe by Total Quantity Sold in descending order
df = df.sort_values("Total Quantity Sold", ascending=False)

# plot the bar plot
fig, ax = plt.subplots(figsize=(15,8))
ax.bar(df["Product Name"], df["Total Quantity Sold"])
ax.set_xlabel("Product Name")
ax.set_ylabel("Total Quantity Sold")

# rotate the x-axis labels to make them visible
plt.xticks(rotation=90)

# adjust the size of the plot to make the product names more visible
plt.tight_layout()

# show the plot
plt.show()
```

```python
# sort the dataframe by Total Quantity Sold in descending order
df = df.sort_values("Total Sales", ascending=False)

# calculate the cumulative sum of Total Quantity Sold
df["Cumulative Sum"] = df["Total Sales"].cumsum()

# calculate the cumulative sum as a percentage of the total sum
df["Cumulative Sum Percentage"] = df["Cumulative Sum"] / df["Total Sales"].sum() * 100

# select the top 20% of the products based on Total Quantity Sold
top_products = df[df["Cumulative Sum Percentage"] <= 80]

# get the product names of the top products
product_names = top_products["Product Name"].tolist()

# print the recommended product names
print("Recommended Product Names:", product_names)
```

```python
plt.figure(figsize=(20, 12))

# sort the dataframe by Total Quantity Sold in descending order
df = df.sort_values("Total Sales", ascending=False)

# calculate the cumulative sum of Total Quantity Sold
df["Cumulative Sum"] = df["Total Sales"].cumsum()

# calculate the cumulative sum as a percentage of the total sum
df["Cumulative Sum Percentage"] = df["Cumulative Sum"] / df["Total Sales"].sum() * 100

# plot the bar chart of Total Quantity Sold
plt.bar(df["Product Name"], df["Total Sales"], color="blue")

# plot the line chart of Cumulative Sum Percentage
plt.plot(df["Product Name"], df["Cumulative Sum Percentage"], color="red", linestyle="dashed")

# add labels and title to the chart
plt.xlabel("Product Name")
plt.ylabel("Total Sales (Blue) and Cumulative Sum Percentage (Red)")
plt.title("Pareto Chart of Total Sales and Cumulative Sum Percentage")

# rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)

# show the chart
plt.show()
```
